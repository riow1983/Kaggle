```python
df = pd.read_csv("./train.csv")
df.tail()
```

```python
uniqcols = ["Name", "Ticket", "Cabin"]
```

```python
df.drop(columns=uniqcols, inplace=True)
```

```python
df.set_index("PassengerId", inplace=True)
```

```python
df["Sex"] = df["Sex"].apply(lambda x: 0 if x=="female" else 1)
```

```python
# make Embarked to one-hot vector
```

```python
df = pd.concat([df, pd.get_dummies(df["Embarked"])], axis=1)
```

```python
df.drop("Embarked", axis=1, inplace=True)
```

```python
df.head()
```

```python
scaler = StandardScaler()
```

```python
dfx = pd.DataFrame(scaler.fit_transform(df.iloc[:, 1:]), columns=df.columns[1:], index=df.index)
```

```python
dfy = df[["Survived"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfx.values, dfy.values, test_size=0.33, random_state=42)
X_train
```

```python
X_train.shape
```

```python
#y_train = y_train.reshape(-1)
```

```python
y_train.shape
```

```python
# create model
model = Sequential()
model.add(Dense(input_dim=X_train.shape[1], units=4,
                kernel_initializer='normal', bias_initializer='zeros'))
model.add(Activation('relu'))

for i in range(0, 15):
    model.add(Dense(units=4, 
                    kernel_initializer='normal', bias_initializer='zeros'))
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
model.add(Dense(units=1))
model.add(Activation('softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
