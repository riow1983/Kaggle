```python
h2c = dict()
for h in hl.Holiday.unique():
    ratios = []
    for c in cal.event_name_1.unique()[1:]:
        ratio = fuzz.ratio(h.lower(), c.lower())
        #print(h, ": ", c, "; ", ratio)
        ratios.append(ratio)
    most_similar_ind = np.argmax(ratios)
    most_similar_string = cal.event_name_1.unique()[1:][most_similar_ind]
    #print(h, most_similar_string)
    if ratios[most_similar_ind] > 50:
        h2c[h] = [most_similar_string, ratios[most_similar_ind]]
    else:
        h2c[h] = []
```

```python
nba_finals_dates = [
    "2011-05-31", "2011-06-02", "2011-06-05", "2011-06-07", "2011-06-09", "2011-06-12",
    "2012-06-12", "2012-06-14", "2012-06-17", "2012-06-19", "2012-06-21", 
    "2013-06-06", "2013-06-09", "2013-06-11", "2013-06-13", "2013-06-16", "2013-06-18", "2013-06-20",
    "2014-06-05", "2014-06-08", "2014-06-10", "2014-06-12", "2014-06-15",
    "2015-06-04", "2015-06-07", "2015-06-09", "2015-06-11", "2015-06-14", "2015-06-16",
    "2016-06-02", "2016-06-05", "2016-06-08", "2016-06-10", "2016-06-13", "2016-06-16", "2016-06-19",
]

pentecost_dates = [
    "2011-06-12", "2012-05-27", "2013-05-19", "2014-06-08",
    "2015-05-24", "2016-05-15",
]

orthodox_pentecost_dates = [
    "2011-06-12", "2012-06-03", "2013-06-23", "2014-06-08",
    "2015-05-31", "2016-06-19",
]
```

```python
nba_finals_startend = cal[cal["event_name_1"].str.contains("NBA")==True].date.tolist()
```

```python
def add_event(df, event_name, event_type, event_dates=[], nba_flag=0):
    df[event_name] = np.nan
    df["event_name_3"] = np.nan
    df["event_type_3"] = np.nan
    df["event_name_4"] = np.nan
    df["event_type_4"] = np.nan
    df["event_name_5"] = np.nan
    df["event_type_5"] = np.nan
    for i,date in enumerate(df.date):
        #print(i, date)
        if date in event_dates:
            if nba_flag==1:
                if date not in nba_finals_startend:
                    df.loc[i, event_name] = 1
                else:
                    pass
            else:
                df.loc[i, event_name] = 1
        else:
            pass
    for i,event in enumerate(df[event_name]):
        #print(i, event)
        if event == 1:
            if df.loc[i, "event_name_1"] != df.loc[i, "event_name_1"]:
                df.loc[i, "event_name_1"] = event_name
                df.loc[i, "event_type_1"] = event_type
            elif df.loc[i, "event_name_2"] != df.loc[i, "event_name_2"]:
                df.loc[i, "event_name_2"] = event_name
                df.loc[i, "event_type_2"] = event_type
            elif df.loc[i, "event_name_3"] != df.loc[i, "event_name_3"]:
                df.loc[i, "event_name_3"] = event_name
                df.loc[i, "event_type_3"] = event_type
            elif df.loc[i, "event_name_4"] != df.loc[i, "event_name_4"]:
                df.loc[i, "event_name_4"] = event_name
                df.loc[i, "event_type_4"] = event_type
            elif df.loc[i, "event_name_5"] != df.loc[i, "event_name_5"]:
                df.loc[i, "event_name_5"] = event_name
                df.loc[i, "event_type_5"] = event_type
            else:
                print(df.loc[i, "date"], "needs event_name_6!")
        else:
            pass
    df.drop(event_name, axis=1, inplace=True)
    return df
```
